#Define Classes

```python
# Student database class
class StudentDB:
  def __init__(self):
    # Import packages and visualization and statistics
    import pandas as pd
    from IPython.display import display_html
    import seaborn as sns
    import matplotlib.pyplot as plt
    import numpy as np
    from scipy.stats import pearsonr

    # Declare datasets
    self.ds1 = None
    self.ds2 = None
    self.ds1removed = None
    self.ds2removed = None
    self.ds1renamed = None
    self.ds2renamed = None
    self.ds_merged = None
    self.display_html = display_html
    self.plt = plt
    self.pearsonr = pearsonr
    self.sns = sns
    self.np = np
    self.pd = pd

  # Read dataset data
  def read_csv(self,url1, url2):
    self.ds1 = pd.read_csv(url1)
    self.ds2 = pd.read_csv(url2)
    self.display_html(self.ds1.head())
    print('\n\n')
    self.display_html(self.ds2.head())

  # Remove all columns that we do not want to include in analysis from each dataset
  def remove_columns(self, list1, list2):
    # Dataset 1
    column_names = self.ds1.columns.tolist()
    print(column_names)
    self.ds1removed = self.ds1.drop(list1, axis=1)
    column_names = self.ds1removed.columns.tolist()
    print(column_names)

    # Dataset 2
    column_names = self.ds2.columns.tolist()
    print(column_names)
    self.ds2removed = self.ds2.drop(list2, axis=1)
    column_names = self.ds2removed.columns.tolist()
    print(column_names)

  def rename_columns(self, dict1, dict2):
    self.ds1renamed = self.ds1removed.rename(columns = dict1)
    self.ds2renamed = self.ds2removed.rename(columns = dict2)

  def merge_datasets(self):
    self.ds_merged = pd.concat([self.ds1renamed, self.ds2renamed])
    self.display_html(self.ds_merged.head())

  def scatter_plot(self, x, y):
    # Create 3 side-by-side scatter plots calculating correlation coefficient
    fig, axes = self.plt.subplots(1, 3, figsize=(18, 5))

    # Calculate the correlation coefficient 1
    r1_value, p1_value = self.pearsonr(x[0], y[0])

    # Calculate the correlation coefficient 2
    r2_value, p2_value = self.pearsonr(x[1], y[1])

    # Calculate the correlation coefficient 3
    r3_value, p3_value = self.pearsonr(x[2], y[2])

    axes[0].scatter(x[0], y[0], color='blue', label='Scatter Plot 1', s=4)
    axes[0].set_title(f"Correlation Coefficient (R = {r1_value:.2f})")
    axes[0].set_xlabel(x[0].name)
    axes[0].set_ylabel(y[0].name)

    axes[1].scatter(x[1], y[1], color='purple', label='Scatter Plot 2', s=4)
    axes[1].set_title(f"Correlation Coefficient (R = {r2_value:.2f})")
    axes[1].set_xlabel(x[1].name)
    axes[1].set_ylabel(y[1].name)

    axes[2].scatter(x[2], y[2], color='orange', label='Scatter Plot 3', s=4)
    axes[2].set_title(f"Correlation Coefficient (R = {r3_value:.2f})")
    axes[2].set_xlabel(x[2].name)
    axes[2].set_ylabel(y[2].name)

    self.plt.tight_layout
    self.plt.show()

  def box_plot(self, x, y):
    # Create 3 boxplots
    self.sns.catplot(x=x[0], y=y[0], data=self.ds_merged, kind='box', color="lightblue")
    self.plt.title(x[0] + ' and ' + y[0])
    self.plt.xlabel(x[0])
    self.plt.ylabel(y[0])
    self.plt.show()

    self.sns.catplot(x=x[1], y=y[1], data=self.ds_merged, kind='box', color="lightgreen")
    self.plt.title(x[1] + ' and ' + y[1])
    self.plt.xlabel(x[1])
    self.plt.ylabel(y[1])
    self.plt.show()

    self.sns.catplot(x=x[2], y=y[2], data=self.ds_merged, kind='box', color="gold")
    self.plt.title(x[2] + ' and ' + y[2])
    self.plt.xlabel(x[2])
    self.plt.ylabel(y[2])
    self.plt.show()
```

#Formantion of a Dataset for Analysis

```python
# Create a Student DB object
student_db = StudentDB()
```

```python
# Define path to dataset 1
url1 = 'https://raw.githubusercontent.com/CS22B-IntroPython/mini-assignments-spring2025-es/main/Project/DataSet1.csv'

# Define path to dataset 2
url2 = 'https://raw.githubusercontent.com/CS22B-IntroPython/mini-assignments-spring2025-es/main/Project/DataSet2.csv'

# Read in datasets
student_db.read_csv(url1, url2)
```

```python
# Make the datasets homogeneous

# Remove unnecessary columns
columns1_to_remove = ['Teacher_Quality','School_Type','Peer_Influence','Learning_Disabilities', 'Parental_Education_Level', 'Gender']
columns2_to_remove = ['Student_ID','First_Name','Last_Name','Email', 'Age', 'Department', 'Midterm_Score', 'Grade', 'Stress_Level (1-10)', 'Parent_Education_Level', 'Gender']
student_db.remove_columns(columns1_to_remove, columns2_to_remove)
```

```python
# Merge the two datasets together
student_db.merge_datasets()
```

#Calculate a Feature Importance Matrix

```python
# Convert categorical data to numerical using One Hot Encoding
from sklearn.preprocessing import OneHotEncoder
import re

# Before the Matrix can be calculated, all categorical comuns need to be converted to ml=utiple nummerical columns
df = student_db.ds_merged.copy()
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # handle_unknown='ignore' to avoid errors with unseen values during prediction

one_hot_encoded = encoder.fit_transform(df[categorical_columns])

# Get feature names and make them unique
feature_names = encoder.get_feature_names_out(categorical_columns)

one_hot_df = pd.DataFrame(one_hot_encoded, columns=feature_names, index=df.index) # Use original feature names and align index with df


DataSetEncoded = pd.concat([df, one_hot_df], axis=1)

DataSetEncoded = DataSetEncoded.drop(categorical_columns, axis=1)

# Need to drop columns containing "_nan")
pattern = r'_nan'

# Identify columns matching the pattern
columns_to_drop = [col for col in DataSetEncoded.columns if re.search(pattern, str(col))]

# Drop the identified columns
DataSetEncoded = DataSetEncoded.drop(columns=columns_to_drop, axis=1)
```

```python
# Import necessary library
from sklearn.ensemble import RandomForestRegressor

# Separate features (X) and target variable (y)
x = DataSetEncoded.drop('Final_Score', axis=1)
y = DataSetEncoded['Final_Score']

# Train a Random Forest Regressor model
model = RandomForestRegressor()
model.fit(x, y)

# Get feature importances
importances = model.feature_importances_

# Create a DataFrame for feature importances
feature_importance_df = pd.DataFrame({'Feature': x.columns, 'Importance': importances})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print the feature importance matrix
print(feature_importance_df)

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.bar(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importance Matrix')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#Calculate Correlation Matrix and VIF (Variance Inflation Factor)

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
corr_matrix = DataSetEncoded.corr(numeric_only=True)

fig, ax = plt.subplots(figsize=(13,13))

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"fontsize":7})

# Set the title
plt.title('Correlation Matrix')
```

```python
# Before we can run the VIF matrix calculation, we need to normalize the data

from sklearn.preprocessing import MinMaxScaler

df = DataSetEncoded.copy()
df.drop('Final_Score', axis=1, inplace=True)
columns_to_normalize = df.columns
scaler = MinMaxScaler()
DataSetEncoded[columns_to_normalize] = scaler.fit_transform(DataSetEncoded[columns_to_normalize])
```

```python
# Import command to suppress warnings
import warnings
warnings.filterwarnings('ignore', category=RuntimeWarning)

# Import necessary libraries
from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

X = DataSetEncoded.select_dtypes(include=['number']).drop(columns=['Final_Score'], errors='ignore')

# Add a constant to the model (intercept)
X = add_constant(X)

for col in X.columns:
  # Check if the column contains infinite values
  if np.isinf(X[col]).any():
    # Replace infinite values with NaN
    X[col] = X[col].replace([np.inf, -np.inf], np.nan)
  if np.isnan(X[col]).any():
    # Impute NaN values with the column mean
    X[col] = X[col].fillna(X[col].mean())

vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns

# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

print(vif_data)
```

```python
# Removing all highly correlated variables

while vif_data['VIF'].max() > 5:
  row_index = vif_data['VIF'].idxmax()
  column_name = vif_data.loc[row_index, 'Feature']
  X = X.drop(column_name, axis=1)
  vif_data = pd.DataFrame()
  vif_data["Feature"] = X.columns
  vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
print('\n\n')
print(vif_data)
```

#Feature Importance Based on Independent Features

```python
# Separate features (X) and target variable (y)
x = DataSetEncoded[(selected_columns for selected_columns in vif_data['Feature'])].copy()
y = DataSetEncoded['Final_Score']

# Train a Random Forest Regressor model
model = RandomForestRegressor()
model.fit(x, y)

# Get feature importances
importances = model.feature_importances_

# Create a DataFrame for feature importances
feature_importance_df = pd.DataFrame({'Feature': x.columns, 'Importance': importances})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print the feature importance matrix
print(feature_importance_df)

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.bar(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Truly Independent Feature Importance Matrix')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
