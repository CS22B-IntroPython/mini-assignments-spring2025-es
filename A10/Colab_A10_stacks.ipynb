## Question 1: Write a function to sort a stack.

```python
def sort_stack(stack):
  sorted = []
  while stack:
        temp = stack.pop()
        while sorted and sorted[-1] > temp:
            stack.append(sorted.pop())
        sorted.append(temp)
  return sorted

stack = [34, 3, 31, 98, 92, 23]
print(sort_stack(stack))
```

2. Create a function that sums all of the "children". Test with the node 10 (Expected answer is 37).


```python
tree = [8,[3,[1,None,None],[6,[4,None,None],[7,None,None]]],[10,None,[14,[13,None,None],None]]]
print(tree)
```

```python
def tree_sum(node):
  sum = node[0]
  if (node[1] != None):
   sum = sum + tree_sum(node[1])
  if (node[2] != None):
   sum = sum + tree_sum(node[2])

  return sum

print(tree_sum(tree[2]))
```
